## Part 1 – Initial description of the datasets

**UN Population dataset (Excel):**
- Contains demographic indicators by country and year (e.g., population, births, deaths, net migrants).
- Year coverage is expected to be roughly 1950–2023 (based on the specification).

**Primary School Enrolments dataset (CSV):**
- Contains primary school enrolment numbers by country and year.

## Part 1 – Subsetting the required countries

```python
countries = ["Malaysia", "United Arab Emirates", "United Kingdom"]

df_un = df_un_raw[
    df_un_raw["Region, subregion, country or area *"].isin(countries)
].copy()

df_un.head()
```

```python
df_un["Region, subregion, country or area *"].unique()
```

```python
df_enrol = df_enrol_raw[
    df_enrol_raw["Country Name"].isin(countries)
].copy()

df_enrol.head()
```

```python
df_enrol["Country Name"].unique()
```

## Part 1 – Cleaning and renaming columns

The original datasets contain long column names and unnecessary metadata.
To make analysis easier and clearer, I select only the required columns and rename them into simpler, consistent variable names.

```python
# select only useful columns
df_un = df_un[[
    "Region, subregion, country or area *",
    "Year",
    "Total Population, as of 1 July (thousands)",
    "Births (thousands)",
    "Total Deaths (thousands)",
    "Net Number of Migrants (thousands)"
]].copy()

# rename columns to clean names
df_un = df_un.rename(columns={
    "Region, subregion, country or area *": "country",
    "Total Population, as of 1 July (thousands)": "population_thousands",
    "Births (thousands)": "births_thousands",
    "Total Deaths (thousands)": "deaths_thousands",
    "Net Number of Migrants (thousands)": "net_migrants_thousands"
})

df_un.head()
```

```python
df_un.columns
```

## Converting units

The UN dataset provides demographic indicators in thousands.
To ensure consistency with the enrolment dataset (which is in actual counts), 
I convert all demographic variables from thousands into actual population numbers.

```python
# convert thousands to actual counts
df_un["population"] = df_un["population_thousands"] * 1000
df_un["births"] = df_un["births_thousands"] * 1000
df_un["deaths"] = df_un["deaths_thousands"] * 1000
df_un["net_migrants"] = df_un["net_migrants_thousands"] * 1000

# drop old thousand columns
df_un = df_un.drop(columns=[
    "population_thousands",
    "births_thousands",
    "deaths_thousands",
    "net_migrants_thousands"
])

df_un.head()
```

```python
# --- Fix dtypes: make sure numeric columns are numeric (not object) ---
df_un["Year"] = pd.to_numeric(df_un["Year"], errors="coerce").astype("Int64")

num_cols = ["population", "births", "deaths", "net_migrants"]
for c in num_cols:
    df_un[c] = pd.to_numeric(df_un[c], errors="coerce")

print(df_un[num_cols].isna().sum())     # check if coercion created NaNs
print(df_un.dtypes)                     # confirm dtypes
df_un.head()
```

```python
df_un.columns
```

## Part 1 – Reshaping enrolment data (wide → long format)

The enrolment dataset is currently stored in a wide format,
where each year is a separate column.

To make analysis easier and consistent with tidy data principles,
I reshape it into a long format where each row represents:

(country, year, enrolment)

```python
# identify year columns (they are numeric strings like '1960', '1961', ...)
year_columns = [col for col in df_enrol.columns if col.isdigit()]

# reshape into long format
df_enrol_long = df_enrol.melt(
    id_vars=["Country Name"],
    value_vars=year_columns,
    var_name="Year",
    value_name="enrolment"
)

# rename country column for consistency
df_enrol_long = df_enrol_long.rename(columns={
    "Country Name": "country"
})

df_enrol_long.head()
```

```python
df_enrol_long.head(10)
```

```python
# convert data types
df_enrol_long["Year"] = df_enrol_long["Year"].astype(int)
df_enrol_long["enrolment"] = pd.to_numeric(df_enrol_long["enrolment"], errors="coerce")

df_enrol_long.dtypes
```

Year is converted to integer type and enrolment values are converted to numeric.
Any non-numeric entries are coerced to NaN to properly handle missing data.

## Handling missing values and checking year coverage

Enrolment values contain missing entries (NaN), especially in early years.
I inspect the amount of missing data and identify the usable year range for analysis.

```python
df_enrol_long["enrolment"].isna().sum(), df_enrol_long.shape[0]
```

```python
df_enrol_long.groupby("country")["enrolment"].apply(lambda s: s.notna().sum())
```

```python
def first_last_valid_year(sub):
    valid = sub.dropna(subset=["enrolment"])
    return pd.Series({
        "first_year_with_data": valid["Year"].min(),
        "last_year_with_data": valid["Year"].max()
    })

df_enrol_long.groupby("country")[["Year", "enrolment"]].apply(first_last_valid_year)
```

## Merging datasets

After cleaning and reshaping, I merge the UN demographic dataset with the enrolment dataset
using the keys (country, Year). This produces a single tidy table for analysis and plotting.

```python
df_un["Year"] = df_un["Year"].astype(int)
df_un.dtypes
```

```python
df_merged = pd.merge(
    df_un,
    df_enrol_long,
    on=["country", "Year"],
    how="left"
)

df_merged.head(10)
```

```python
df_merged.groupby("country").size()
```

## Feature: Natural change

Natural change is defined as births minus deaths.  
I compute it to support comparison with population change and net migration.

```python
df_merged["natural_change"] = df_merged["births"] - df_merged["deaths"]
df_merged[["country", "Year", "births", "deaths", "natural_change"]].head()
```

## Feature Engineering: Eligible Primary Age Population

The UN dataset does not directly provide the number of children of primary school age.

Based on World Bank and demographic structure references, primary school age children 
(approximately ages 6–11) typically account for around 12% of the total population.

Therefore, I estimate the eligible primary school age population as:

Eligible primary age population = 12% × total population

This estimation is based on demographic structure references suggesting that children aged 6–11 typically account for approximately 12% of the total population in many countries.

However, since this is an approximation and age structure varies across countries and over time, a sensitivity analysis using 10%, 12% and 14% of total population is conducted.

This allows us to assess whether conclusions remain robust under different reasonable demographic assumptions.

Population structure estimates are informed by general demographic patterns reported by the World Bank and UN population statistics.

```python
# --- Sensitivity check for eligible primary age population assumption ---
# If 12% is an approximation, check whether conclusions hold for 10%-14% too.
df_merged = df_merged.copy()

df_merged["eligible_10pct"] = df_merged["population"] * 0.10
df_merged["eligible_12pct"] = df_merged["population"] * 0.12
df_merged["eligible_14pct"] = df_merged["population"] * 0.14

df_merged[["country","Year","population","eligible_10pct","eligible_12pct","eligible_14pct"]].head()
```

```python
df_merged["primary_age_population"] = df_merged["population"] * 0.12

df_merged[[
    "country","Year","population","primary_age_population"
]].head()
```

```python
df_merged[[
    "population",
    "births",
    "deaths",
    "natural_change",
    "net_migrants",
    "primary_age_population",
    "enrolment"
]].describe()
```

The descriptive statistics reveal substantial differences in demographic scale across the three countries. 
The United Kingdom maintains the highest overall population levels, while Malaysia shows steady demographic expansion over time. 
The UAE exhibits relatively larger fluctuations in migration-related indicators, reflecting its reliance on international labour mobility. 
Primary school enrolment figures broadly align with the size of the eligible primary age population across the nations.

## Part 2: Visualisation and Analysis

This section presents visualisations and brief analytical insights comparing demographic trends across Malaysia, the United Arab Emirates and the United Kingdom. The graphs illustrate population growth, natural change, migration trends and the relationship between eligible primary school age population and actual primary school enrolment.

## Primary school enrolment over time

This chart shows how primary school enrolment has changed over time
in Malaysia, the UAE and the UK.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

for c in df_merged["country"].unique():
    sub = df_merged[df_merged["country"]==c]
    plt.plot(sub["Year"], sub["enrolment"], label=c)

plt.title("Primary School Enrolment Over Time")
plt.xlabel("Year")
plt.ylabel("Number of pupils")
plt.legend()
plt.show()
```

Lines are discontinuous because enrolment values are missing (NaN) for some years in the World Bank dataset, so Matplotlib does not connect points across gaps.

## Population vs enrolment

This chart compares total population with primary school enrolment
to observe whether enrolment follows demographic trends.

```python
plt.figure(figsize=(10,6))

for c in df_merged["country"].unique():
    sub = df_merged[df_merged["country"]==c]
    plt.plot(sub["Year"], sub["population"], label=c)

plt.title("Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
df_growth = df_merged.sort_values(["country", "Year"]).copy()
df_growth["pop_growth_rate_pct"] = df_growth.groupby("country")["population"].pct_change() * 100

plt.figure(figsize=(12,6))
for c in countries:
    temp = df_growth[df_growth["country"] == c]
    plt.plot(temp["Year"], temp["pop_growth_rate_pct"], label=c)

plt.title("Annual Population Growth Rate (%)")
plt.xlabel("Year")
plt.ylabel("Growth rate (%)")
plt.legend()
plt.grid(True)
plt.show()

df_growth[["country","Year","population","pop_growth_rate_pct"]].head(10)
```

## Net migration trends

Net migration can influence population growth and school enrolment,
especially in countries with high inward migration such as the UAE.

```python
plt.figure(figsize=(10,6))

for c in df_merged["country"].unique():
    sub = df_merged[df_merged["country"]==c]
    plt.plot(sub["Year"], sub["net_migrants"], label=c)

plt.title("Net Migration Over Time")
plt.xlabel("Year")
plt.ylabel("Net migrants")
plt.legend()
plt.show()
```

## Population change vs births, natural change and net migration

To compare the drivers of population growth, I compute the annual population change 
(current year population minus previous year population) and compare it with births, 
natural change and net migration.

```python
# calculate yearly population change
df_merged = df_merged.sort_values(["country", "Year"]).copy()
df_merged["population_change"] = df_merged.groupby("country")["population"].diff()

# plot comparison for each country
for c in df_merged["country"].unique():
    sub = df_merged[df_merged["country"] == c].dropna(subset=["population_change"]).copy()

    plt.figure(figsize=(10,6))
    plt.plot(sub["Year"], sub["population_change"], label="Population change")
    plt.plot(sub["Year"], sub["births"], label="Births")
    plt.plot(sub["Year"], sub["natural_change"], label="Natural change")
    plt.plot(sub["Year"], sub["net_migrants"], label="Net migration")

    plt.title(f"{c}: Population change vs births, natural change and net migration")
    plt.xlabel("Year")
    plt.ylabel("People")
    plt.legend()
    plt.show()
```

## Discussion of findings

The analysis shows clear demographic differences between Malaysia, the UAE and the UK.

Malaysia demonstrates steady population growth and increasing primary school enrolment over time, reflecting sustained natural population growth. The UK shows relatively stable population growth and enrolment trends, consistent with a mature demographic structure.

The UAE displays more fluctuation in both population and migration trends. This is expected due to its high dependence on international migration, which significantly affects population size and school enrolment demand.

Overall, demographic factors such as population growth, natural change and migration appear to influence primary school enrolment trends. Countries with higher population growth generally show increasing enrolment numbers, while migration plays a particularly strong role in shaping enrolment in the UAE.

These findings are consistent with general demographic patterns reported by the World Bank and UN population statistics, where population structure and migration significantly affect education demand.

In absolute terms, the UAE experienced the most rapid relative population expansion over the observed decades, largely driven by migration rather than natural increase. Malaysia’s growth appears more structurally demographic (births exceeding deaths), while the UK demonstrates slower but stable expansion with demographic ageing influences.

## Eligible Primary Age Population vs Actual Enrolment

This chart compares the estimated eligible primary school age population 
with the actual primary school enrolment numbers for the UK, UAE and Malaysia.

This allows us to observe whether enrolment trends follow demographic trends.

```python
for c in df_merged["country"].unique():
    temp = df_merged[df_merged["country"] == c].copy()
    temp = temp.sort_values("Year")

    plt.figure(figsize=(12,6))

    plt.plot(temp["Year"], temp["enrolment"],
             label="Enrolment (World Bank)", linewidth=2)

    plt.plot(temp["Year"], temp["eligible_10pct"],
             linestyle="--", label="Eligible (10%)")

    plt.plot(temp["Year"], temp["eligible_12pct"],
             linestyle="--", label="Eligible (12%)")

    plt.plot(temp["Year"], temp["eligible_14pct"],
             linestyle="--", label="Eligible (14%)")

    plt.title(f"Enrolment vs Estimated Eligible Primary Age Population - {c}")
    plt.xlabel("Year")
    plt.ylabel("Number of pupils")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## Insights

The results indicate that primary school enrolment trends generally follow changes in the estimated eligible primary school age population across the three countries. Countries experiencing sustained growth in child population tend to show increasing enrolment over time.

However, the relationship is not perfectly proportional. Differences between eligible population and actual enrolment may reflect variations in education policy, migration patterns, demographic structure and access to schooling. For example, countries with high levels of working-age migration may experience rapid total population growth without a corresponding increase in school-age children.

### Country-level observations

Malaysia shows relatively steady demographic growth driven by natural population increase, with enrolment trends broadly following the growth of the eligible population.

The United Arab Emirates demonstrates rapid population expansion influenced strongly by migration. However, enrolment does not always increase at the same rate, reflecting the large proportion of expatriate working-age adults without school-age dependents.

The United Kingdom shows slower demographic growth and periods of stabilisation or decline in enrolment, consistent with long-term fertility trends and demographic transition patterns.

## Overall Conclusion

Overall, while the estimated eligible primary school age population continues to grow in all three countries, actual enrolment trends do not always increase at the same rate. This suggests that demographic growth alone does not fully determine school enrolment patterns, as government policy, migration, and education system capacity also play important roles.

The sensitivity analysis demonstrates that even when different reasonable assumptions (10–14%) are used to estimate the eligible population, the overall trends and interpretations remain consistent. This increases confidence in the robustness of the analysis and supports the validity of the conclusions drawn from the visualisations.
